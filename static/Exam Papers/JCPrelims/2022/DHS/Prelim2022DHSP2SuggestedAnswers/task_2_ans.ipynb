```python
class CircularQueue(): 
 
    # constructor
    def __init__(self, size): # initializing the class

        self.size = size 

        # initializing queue with none
        self.queue = [None for i in range(size)] 
        self.front = self.rear = -1 
    
    def display_raw_array(self):

        print("Fixed array contents:")
        print(self.queue)
        print()
        print()
 

# STUDENT VERSION:

    # def enqueue(self, data):
    #   pass # Replace this with your code
             
    # def dequeue(self):
    #   pass # Replace this with your code


# START OF ANSWER KEY:

    def enqueue(self, string):
         
        # condition if Cannot enqueue: Queue is full.
        if ((self.rear + 1) % self.size == self.front):
            return "Cannot enqueue: Queue is full." # 1m
             
        # condition for empty queue
        elif (self.front == -1):  # 1m
            self.front = 0
            self.rear = 0 # 1m
            self.queue[self.rear] = string # 1m

        else:
            # next position of rear
            self.rear = (self.rear + 1) % self.size # 1m
            self.queue[self.rear] = string # 1m
             

    def dequeue(self):

        if (self.front == -1): # condition for empty queue
            return ("Queue is empty") # 1m
             
        # condition for only one element
        elif (self.front == self.rear): # 1m
            temp=self.queue[self.front]
            self.front = -1
            self.rear = -1
            return temp # 1m

        else:
            temp = self.queue[self.front]
            self.front = (self.front + 1) % self.size
            return temp # 1m

# END OF ANSWER KEY


# Driver program
platform = CircularQueue(5)

print("(1) Enqueue Participant_A, Participant_B, Participant_C, Participant_D.")
print()
return_value = platform.enqueue("Participant_A")
if return_value == "Cannot enqueue: Queue is full.": print(return_value)

return_value = platform.enqueue("Participant_B")
if return_value == "Cannot enqueue: Queue is full.": print(return_value)

return_value = platform.enqueue("Participant_C")
if return_value == "Cannot enqueue: Queue is full.": print(return_value)

return_value = platform.enqueue("Participant_D")
if return_value == "Cannot enqueue: Queue is full.": print(return_value)

platform.display_raw_array()


print ("(2.1) Deleted value = ", platform.dequeue())
print ("(2.2) Deleted value = ", platform.dequeue())
platform.display_raw_array()


print("(3) Enqueue Participant_E, Participant_F, Participant_G.")
print()


return_value = platform.enqueue("Participant_E")
if return_value == "Cannot enqueue: Cannot enqueue: Queue is full..": print(return_value)

return_value = platform.enqueue("Participant_F")
if return_value == "Cannot enqueue: Cannot enqueue: Queue is full..": print(return_value)

return_value = platform.enqueue("Participant_G")
if return_value == "Cannot enqueue: Cannot enqueue: Queue is full..": print(return_value)

platform.display_raw_array()


print("(4) Enqueue Participant_H.")
print()

return_value = platform.enqueue("Participant_H")
if return_value == "Cannot enqueue: Queue is full.": print(return_value)

platform.display_raw_array()


```

```python
def simulate(filename, threshold):

  file = open(filename)

  timings = []

  for item in list(file): 
    timings.append(item.strip().split(',')) # 1m

  file.close()

  time_entry = {}

  for item in timings: 
    time_entry[item[2]] = ""
   # 1m
  

  platform = CircularQueue(threshold) 

  for item in time_entry.keys(): # 1m
    for row in timings:

      if row[2] == item:
 
        if platform.enqueue(row[1]) == "Cannot enqueue: Queue is full.": 
          
          return "Full at " + row[2] + " when " + row[1] + " enters." # 1m

      if row[3] == item:
        platform.dequeue() # 1m

        
 
print(simulate("marching.csv", 530)) 

```
