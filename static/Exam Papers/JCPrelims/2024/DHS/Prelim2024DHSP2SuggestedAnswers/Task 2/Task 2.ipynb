```python
#Task 2.1
class Stack:
    def __init__(self):
        self.maxSize = 15 #1 mark for declaring and initialising maxSize and topPointer variables with the correct values 
        self.stackList = []
        for i in range(self.maxSize): #1 mark for declaring and initialising stackList correctly
            self.stackList.append(None) #populating the list to the number of elements based on maxSize
        self.topPointer = -1 #the topPointer pointer starts from -1 so that when we add the first element, we add 1 to get index 0 which follows the index of the python list ==> less confusing

    def is_full(self): #1 mark for correctly checking whether a stack is full
        if self.topPointer == self.maxSize - 1:
            return True
        else:
            return False

    def push(self,webpage_link):
        if self.is_full():
            for i in range(self.maxSize - 1): #2 mark for shifting elements left by 1 correctly if stack is full - to allow the new web page link to be inserted. Need to reduce topPointer by 1 after shifting
                self.stackList[i] = self.stackList[i+1] #shift everything leftwards by 1
            self.topPointer = self.topPointer - 1
        self.topPointer = self.topPointer + 1 #1 mark for pushing element into the stack correctly
        self.stackList[self.topPointer] = webpage_link #adding new element to the top of the stack (the index of the top of the stack is indicated by the topPointer variable)

    def is_empty(self):  #1 mark for checking correctly whether a stack is empty and returning "History is empty" if stack is empty (for both pop and display method)
        if self.topPointer == -1:
            return True
        else:
            return False

    def pop(self):
        if self.is_empty():
            return "History is empty." 
        else: #1 mark for popping correctly
            popped_item = self.stackList[self.topPointer] #retriving the element/value that the topPointer pointer variable is currently pointing at (Note: there is no need to remove the value in the Stack, just move the topPointer variable down in the next line of code)
            self.topPointer = self.topPointer - 1 #moving the topPointer pointer variable down by one to point to the next element
            return popped_item #return the value of the deleted element

    def display(self): #1 mark for displaying the Stack as stated in the question
        if self.is_empty():
            print("History is empty.")
        else:
            for i in range(self.topPointer, -1, -1):
                if i == self.topPointer:
                    print(f"{self.stackList[i]} <- topPointer")
                else:
                    print(f"{self.stackList[i]}")
```

```python
#Task 2.3
class Node:

    def __init__(self, data): #1 mark for Node and LinkedList constructor with correct instance variables
        self.data = data
        self.next = None

    def display(self):
        print(self.data)

    def get_data(self):
        return self.data

    def get_next(self):
        return self.next

    def set_data(self, data):
        self.data = data

    def set_next(self, next):
        self.next = next
        
class LinkedList:

    def __init__(self): 
        self.head = None
        
    def insert(self, webpage_link): #add_to_front
        new_node = Node(webpage_link)
        if self.head ==  None: #1 mark for Scenario 1 - empty Linked List
            self.head = new_node 
        else: #1 mark for Scenario 2 - not empty Linked List
            new_node.set_next(self.head) #pointing the new node to the current head node
            self.head = new_node #pointing the head pointer to the new node
            
    def delete(self, webpage_link): #delete_by_value
        if self.head == None: #1 mark for Scenario 1 - Linked List is empty and Scenario 4 - the value is not found
            print("Not found")
            return None
        elif self.head.get_data() == webpage_link: #1 mark for Scenario 2 - the node to delete is the current head node
            node_to_delete = self.head
            self.head = self.head.get_next() #head pointer will point to the next node to be the new head node
            return node_to_delete
        else: # 1 mark for Scenario 3 - the node to delete is from the 2nd node onwards including last node
            previous_node = self.head

            while previous_node.get_next() != None: #iterate through the nodes
                current_node = previous_node.get_next()
                if current_node.get_data() == webpage_link: #checking if node is the one to be deleted
                    previous_node.set_next(current_node.get_next()) #point the previous node to the node after the current node (current node is the one to be deleted)
                    return current_node
                else:
                    previous_node = current_node
            #print("Not found") Scenario 4 - the value is not found
            return None
        
    def revisit(self, webpage_link):
        revisit_node = self.delete(webpage_link) # 1 mark for removing the revisited node from its position in the linked list
        if revisit_node == None:
            print('Page Not Found')
        else:
            self.insert(revisit_node.get_data()) # 1 mark inserting the revisted node to the head of the linked list
        
    def display(self): #1 mark for displaying the Linked List as stated in the question
        if self.head == None:
            print("History is empty")
        else:
            curr_node = self.head
            while curr_node != None:
                curr_node.display()
                curr_node = curr_node.next
```
