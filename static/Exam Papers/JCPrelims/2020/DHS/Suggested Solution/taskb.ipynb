```
class Stack:

  def __init__(self):
    self.data = []
    self.top = -1

  def is_empty(self):
    return self.top == -1

  def push(self, value):
    self.data.append(value)
    self.top += 1

  def pop(self):
    if self.is_empty():
      return "Cannot delete from empty stack"
    else:
      temp = self.data[self.top]
      del self.data[self.top]
      self.top -= 1
      return temp

  def show(self):
    if self.is_empty():
      print("Empty stack", end='')
    for i in range(self.top, -1, -1):
      print(self.data[i], end=' ')
    print()


# main
expr = "231+*5-"
stack = Stack()
for token in expr:
  if token.isdigit(): # operand
    stack.push(int(token))
  else: # operator
    data2 = stack.pop()
    data1 = stack.pop()
    if token == '+':
      temp = data1 + data2
    elif token == '-':
      temp = data1 - data2  
    elif token == '*':
      temp = data1 * data2  
    else: # token is /
      temp = data1 / data2        
    stack.push(temp)
print(stack.pop())
#stack.show()

#stack = Stack()
#stack.push(1)
#stack.push(2)
#stack.push(3)
#stack.show()
#print(stack.pop())
#print(stack.pop())
#print(stack.pop())
#print(stack.pop())
#stack.show()
```
