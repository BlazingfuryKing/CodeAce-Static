```python
# Task 2.1

width = input("Please input width of field: ")
width = int(width)
height = input("Please input height of field: ")
height = int(height)

field  = []
for i in range(height):
    lst = []
    for j in range(width):
        lst.append(".")
    field.append(lst)

num_mines = input("Please input number of mines: ")
while True:
    try:
        num_mines = int(num_mines)
    except:
        print("The number of mines must be a positive integer less than width*height.")
        num_mines = input("Please input number of mines: ")
    else:
        if num_mines <= 0 or num_mines >= width*height:
            print("The number of mines must be a positive integer less than width*height.")
            num_mines = input("Please input number of mines: ")
        else:
            break

print("Generating field...") # This line is optional

from random import randint

for k in range(num_mines):
    while True:
        i = randint(0, height-1)
        j = randint(0, width-1)
        if field[i][j] != 'X':
            field[i][j] = 'X'
            break

# 1 mark: input width and height, convert to int
# 1 mark: create field
# 1 mark: input number of mines
# 1 mark: data validation (range of integer)
# 1 mark: import randint
# 1 mark: planting the mines in the field
# 1 mark: ensure that two mines don't occupy the same coordinates

# Total: 7 marks
```

```python
# Task 2.2

def count_mines(field,i,j):
    count = 0
    
    height = len(field)
    width = len(field[0])
    
    if i > 0:
        if j > 0:
            if field[i-1][j-1] == 'X':
                count += 1
        if field[i-1][j] == 'X':
            count += 1
        if j < width - 1:
            if field[i-1][j+1] == 'X':
                count += 1
    if j > 0:
        if field[i][j-1] == 'X':
            count += 1
    if j < width - 1:
        if field[i][j+1] == 'X':
            count += 1
    if i < height - 1:
        if j > 0:
            if field[i+1][j-1] == 'X':
                count += 1
        if field[i+1][j] == 'X':
            count += 1
        if j < width - 1:
            if field[i+1][j+1] == 'X':
                count += 1
    return count

def print_field(field):
    for row in field:
        line = ''
        for cell in row:
            if cell == 'X':
                line += '.'
            else:
                line += cell
        print(line)

        
tries = width*height - num_mines

while True:
    print_field(field)
    if tries == 0:        
        print("Congratulations! You have won!")
        break
    row = int(input("Please input row number to try: "))
    col = int(input("Please input column number to try: "))
    if field[row][col] == 'X':
        print("Sorry! You have stepped on a mine!")
        break
    else:
        display = count_mines(field,row,col)
        field[row][col] = str(display)
        tries -= 1

# 1 mark: counting function/variable
# 1 mark: iterate through surrounding squares
# 1 mark: taking care of squares at the edges
# 1 mark: printing: iterating through field
# 1 mark: printing: . or number as appropriate
# 1 mark: keep track of number of tries or unrevealed squares
# 1 mark: loop
# 1 mark: input row and column number
# 1 mark: stepped on a mine message and end game
# 1 mark: reveal number
# 1 mark: win message

# Total: 11 marks
```
