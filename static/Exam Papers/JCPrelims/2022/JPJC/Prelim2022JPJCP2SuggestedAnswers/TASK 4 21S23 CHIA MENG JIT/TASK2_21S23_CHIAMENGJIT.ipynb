```python
#task 2.1
class Item():
    def __init__(self, cataloguenumber = -1, brand = "", style = "", colour = "", rental = -1): #parameters auto set as null values
        self.CatalogueNumber = cataloguenumber
        self.Brand = brand
        self.Style = style
        self.Colour = colour
        self.Rental = rental
    
    def getCatalogueNumber(self):
        return self.CatalogueNumber
    
    def setCatalogueNumber(self, cataloguenumber):
        self.CatalogueNumber = cataloguenumber
        
    def getBrand(self):
        return self.Brand
    
    def setBrand(self, brand):
        self.Brand = brand
        
    def getStyle(self):
        return self.Style
    
    def setStyle(self, style):
        self.Style = style
        
    def getColour(self):
        return self.Colour
    
    def setColour(self, colour):
        self.Colour = colour
    
    def getRental(self):
        return self.Rental
    
    def setRental(self, rental):
        self.Rental = rental
        
class Apparel(Item):
    def __init__(self, cataloguenumber = -1, brand = "", style = "", colour = "", rental = -1, size = -1, fabrication = ''):
        #above is the null values of the input parameters
        super().__init__(cataloguenumber, brand, style, colour, rental)
        self.Size = size
        self.Fabrication = fabrication
    
    def getSize(self):
        return self.Size
    
    def setSize(self, size):
        self.Size = size
    
    def getFabrication(self):
        return self.Fabrication
    
    def setFabrication(self, fabrication):
        self.Fabrication = fabrication
        

class Bag(Item):
    def __init__(self, cataloguenumber = -1, brand = "", style = "", colour = "", rental = -1, length = -1, breadth = -1, leather = None): 
        #null values of input parameters
        super().__init__(cataloguenumber, brand, style, colour, rental)
        self.Length = length
        self.Breadth = breadth
        self.Leather = leather
    
    def getLength(self):
        return self.Length
    
    def setLength(self, length):
        self.Length = length
    
    def getBreadth(self):
        return self.Breadth
    
    def setBreadth(self, breadth):
        self.Breadth = breadth
    
    def getLeather(self):
        return self.Leather
    
    def setLeather(self, leather):
        self.Leather = leather
    
```

```python
#task 2.2
def Hash(cataloguenumber):
    return cataloguenumber % 100

#====main program=====
hashtable = [() for count in range(0, 100)] #initialises a 0-based array of size 100, () being the null datatype

file = open("RECORDS.csv", 'r') #open file
file.readline() #remove the first line of the file
for line in file:
    cataloguenumber, type, brand, style, colour, rental, size, fabrication, length, breadth, leather, email = line.strip().split(',')
    if type == "Apparel":
        obj = Apparel(int(cataloguenumber), brand, style, colour, int(rental), int(size), fabrication) #obj represents the Apparel object
        #for easier insertion into hashtable later
        hash_address = Hash(int(cataloguenumber))
        record = (obj, email)
        
    elif type == "Bag":
        obj = Bag(int(cataloguenumber), brand, style, colour, int(rental), int(length), int(breadth), leather == "TRUE") #obj represents Bag object for
        #easier insertion into hashtable later
        hash_address = Hash(int(cataloguenumber))
        record = (obj, email)
    

    #===insertion into hashtable========
    #case 1: no collision
    if hashtable[hash_address] == ():
        hashtable[hash_address] = record
    #case 2: linear probing to handle collision
    else:
        cur_pointer = hash_address
        counter = 0 #to count the number of times we traverse through the hashtable
        #print(len(hashtable))
        while True:
            if counter == len(hashtable):
                print("Hash Table is full")
                break
            cur_pointer = (cur_pointer + 1) % len(hashtable)
            counter += 1
            if hashtable[cur_pointer] == ():
                hashtable[cur_pointer] = record
                break

print("Inserted all data into hash table")
file.close()

def display(hashtable, start, end):
    print("{0:10}{1:20}{2:30}".format("Index", "CatalogueNumber", "Email"))
    for index in range(start, end + 1):
        if hashtable[index] == (): #when nothing is in that record
            print("{0:10}{1:20}{2:30}".format(str(index), "None", "None"))
        else:
            print("{0:10}{1:20}{2:30}".format(str(index), str(hashtable[index][0].getCatalogueNumber()), hashtable[index][1]))

display(hashtable, 0, 9)
```

```python
def HashTableSearch(hashtable, CatalogueNumber):
    hash_address = Hash(CatalogueNumber)
    if hashtable[hash_address] == ():
        return "Not Found"
    
    if hashtable[hash_address][0].getCatalogueNumber() == CatalogueNumber:
        return hashtable[hash_address][1]
    
    cur_pointer = hash_address
    counter = 0 #number of times it traverses the hashtable
    while True:
        cur_pointer = (cur_pointer + 1) % len(hashtable)
        counter += 1
        if hashtable[cur_pointer] == ():
            return "Not Found"
        if counter == len(hashtable):
            return "Not Found"
        if hashtable[cur_pointer][0].getCatalogueNumber() == CatalogueNumber:
            return hashtable[cur_pointer][1]
        
#test cases
print(HashTableSearch(hashtable, 1399))
print(HashTableSearch(hashtable, 1220))
```
