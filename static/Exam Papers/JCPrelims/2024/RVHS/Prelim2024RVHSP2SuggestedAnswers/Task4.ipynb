```python
#Task 4.1 [7]
class Contestant:
    def __init__(self, name, time):
        self.name = name
        self.time = time
    
    def get_time(self):
        return self.time
    
    def set_time(self, new_time):
        self.time = new_time
        
    def __str__(self):
        if len(self.name)>= 8:
            first8c = self.name.replace(" ", "")[:8]
            
        else:
            temp = self.name.replace(" ", "")
            first8c = temp + (8-len(temp))*'_'
        return "C-" + first8c.upper() + "("+ str(self.time)+ "s)"

class Beginner(Contestant):
    def __init__(self, name, time, practice_hours):
        super().__init__(name, time)
        self.practice_hours = practice_hours

    def __str__(self):
        return "B" + super().__str__()[1:]
    
class Professional(Contestant):
    
    def __init__(self, name, time, competitions_won):
        super().__init__(name, time)
        self.competitions_won = competitions_won

    def __str__(self):
        return "P" + super().__str__()[1:]

print(str(Contestant("Kan Min Coh", 35.32)))
print(str(Contestant("Man Kin", 15.12)))

print(str(Contestant("ChanChan Chan",35.01)))
print(str(Beginner("Li uLi uLi u",55.54, 300)))
print(str(Professional("OhOhOh",6.34, 60)))
```

```python
#Task 4.2 [3]
def readfile(filename):
    result = []
    f = open(filename, 'r')
    line = f.readline()
    while line:
        kind, name, time, attr = line[:-1].split(',')
        if kind == "Beginner":
            result.append(Beginner(name, float(time), int(attr)))
        else:
            result.append(Professional(name, float(time), int(attr)))
        line= f.readline()
    f.close()
    return result
```

```python
#Task 4.3 [15]
class Node: #[1]
    def __init__(self, player):
        self.player = player # Either a Beginner or a Profrssional instance
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, player): #[3]
        def helper(node):
            if node:
                if node.player.get_time() > player.get_time():
                    if node.left:
                        helper(node.left)
                    else:
                        node.left = Node(player)
                        return
                else:
                    if node.right:
                        helper(node.right)
                    else:
                        node.right = Node(player)
                        return
            else:
                self.root = Node(player)
        helper(self.root)
        
    def inorder(self): #[1]
        def helper(node):
            if node:
                helper(node.left)
                print(str(node.player))
                helper(node.right)
        helper(self.root)
        
    def count_professional(self): #[3]
        def helper(node):
            if node:
                if isinstance(node.player, Professional):
                    return 1 + helper(node.left) + helper(node.right)
                else:
                    return helper(node.left) + helper(node.right)
            else:
                return 0
        return helper(self.root)
    
    def find_best_beginner(self): #[5]
        def helper(node):
            if node:
                left_node = helper(node.left)
                if left_node:
                    return left_node
                else:
                    if isinstance(node.player, Beginner):
                        return str(node.player)
                    else:
                        right_node = helper(node.right)
                        if right_node:
                            return right_node
                        else:
                            return None
            else:
                return None
        return helper(self.root)
    
def testcase(): #[2]
    bst = BST()
    players = readfile("rubik.txt")
    for player in players:
        bst.insert(player)
    bst.inorder()
    print(bst.count_professional())
    print(bst.find_best_beginner())
testcase()
```
