```python
# Task 4.1   
# Create the database and tables (6m)

import sqlite3

f1 = open('SUPERMARKET.db', 'w') #creates the db file
f1.close()

conn = sqlite3.connect('SUPERMARKET.db')

conn.execute('''CREATE TABLE Item (
    ItemID INTEGER PRIMARY KEY, 
    Name TEXT, 
    Price INTEGER);''')

conn.execute('''CREATE TABLE Customer (
    CustomerNumber INTEGER PRIMARY KEY, 
    Name TEXT);''')

conn.execute('''CREATE TABLE Purchase (
    PurchaseID INTEGER PRIMARY KEY, 
    CustomerNumber INTEGER, 
    ItemID INTEGER,
    Quantity INTEGER,
    DateTime TEXT,
    FOREIGN KEY(CustomerNumber) REFERENCES Customer(CustomerNumber),
    FOREIGN KEY(ItemID) REFERENCES Item(ItemID));''')

conn.commit()
conn.close()

# 1m: import sqlite3 and connection to database, incl. commit and close
# 1m: open and close file
# 1m: create Item table
# 1m: create Customer table
# 1m: create Purchase table...
# 1m: ... with correct foreign keys
```

```python
#Task 4.2  
# Populate the 3 tables (5m)

import sqlite3

conn = sqlite3.connect('SUPERMARKET.db')

f1 = open("ITEM.txt")

for line in f1:
    lst = line.strip().split(',')
    number = int(lst[0])
    name = lst[1]
    price = int(lst[2])

    conn.execute('''INSERT INTO Item (ItemID, Name, Price)
                    VALUES(?,?,?)''', (number, name, price))

f1.close()


f2 = open("CUSTOMER.txt")

for line in f2:
    lst = line.strip().split(',')
    number = int(lst[0])
    name = lst[1]

    conn.execute('''INSERT INTO Customer (CustomerNumber, Name)
                    VALUES(?,?)''', (number, name))

f2.close()


f3 = open("PURCHASE.txt")

for line in f3:
    lst = line.strip().split(',')
    pnumber = int(lst[0])
    cnumber = int(lst[1])
    inumber = int(lst[2])
    qty = int(lst[3])
    datetime = lst[4]

    conn.execute('''INSERT INTO Purchase (PurchaseID, CustomerNumber, ItemID, Quantity, Datetime)
                    VALUES(?,?,?,?,?)''', (pnumber, cnumber, inumber, qty, datetime))

f3.close()

conn.commit()
conn.close()

# 1m: import sqlite3 and connection to database, incl. commit and close
# 1m: open and close at least one text file correctly
# 1m: extract data correctly and convert data types (at least one file)
# 1m: insert into at least one table correctly
# 1m: correctly done for all tables
```

```python
# Task 4.3  
# Query for purchases made by customer after a date (6m + 2m)

import sqlite3

num = input("Enter the customer's number: ")      #validation not required
date = input("Enter a date: ")

conn = sqlite3.connect('SUPERMARKET.db')

data = conn.execute('''SELECT Name, Quantity FROM Item INNER JOIN Purchase
                    on Item.ItemID = Purchase.ItemID
                    WHERE CustomerNumber = ?
                    AND DateTime >= ?''', (int(num), date))

lst = data.fetchall()
print("Item Name, Quantity")
for item in lst:
    print(item[0], item[1])

conn.close()

# 1m: import sqlite3 and connection to database, incl. close
# 1m: input
# 1m: query with correct inner join
# 1m: query with correct conditions
# 1m: create list (fetchall)
# 1m: iterate and print items in lst (in this case, none)

# 1m: enter correct inputs
# 1m: printed output (empty)
```
