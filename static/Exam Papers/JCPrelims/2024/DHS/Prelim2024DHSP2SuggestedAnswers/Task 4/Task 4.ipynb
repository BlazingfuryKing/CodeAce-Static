```python
#Task 4.2

# Step 1 - import sqlite3 & csv
import sqlite3
import csv
import datetime

# Step 2 - create database connection
connection = sqlite3.connect("Discipline.db")

with open("Latecoming.csv", "r", newline="") as infile: #1 mark for opening and closing file correctly and for csv reader or other methods of splitting the data

    records = csv.reader(infile, delimiter = ",") #csv reader or other methods of splitting the data
    for record in records: #loop for extracting data for each field
        studentID = record[0]
        name = record[1]
        classname = record[2]
        date_of_latecoming_str = record[3]
        time = record[4]

        # Step 3 - execute SQL query
        cursor = connection.execute("SELECT COUNT(*) FROM Student WHERE StudentID = ?", (studentID,)) # 1 mark for checking whether the student's data exists in the Student Table

        # Step 4 - extracting records
        count = cursor.fetchall()[0][0] #cursor.fetchall() returns [(0,)] when the count is 0
        #print(count)

        if count == 0:
             # Step 3 - execute SQL query
            connection.execute("INSERT INTO Student (StudentID, Name, Class) VALUES (?,?, ?)", (studentID, name, classname)) #1 mark for inserting student's data into Student Table if it does not exist in Student table

        #calculating the Detention date - 1 working day after the latecoming date # 2 marks for calculating the dentention date - if did not consider for Fri, minus 1 mark
        year_of_latecoming = int(date_of_latecoming_str[0:4])

        if date_of_latecoming_str[5] == "0":
            month_of_latecoming = int(date_of_latecoming_str[6])
        else:
            month_of_latecoming = int(date_of_latecoming_str[5:7])

        if date_of_latecoming_str[8] == "0":
            day_of_latecoming = int(date_of_latecoming_str[9])
        else:
            day_of_latecoming = int(date_of_latecoming_str[8:])

        #print(year_of_latecoming)
        #print(month_of_latecoming)
        #print(day_of_latecoming)s

        detention_date = datetime.datetime(year_of_latecoming, month_of_latecoming, day_of_latecoming) + datetime.timedelta(days=1) #calculating detention date by adding 1 day to the latecoming_date
        if detention_date.strftime("%A") == "Saturday":
            detention_date = detention_date + datetime.timedelta(days=2) #adding 2 more days to detention date to become Monday if original detention day is a Saturday
        detention_date_str = detention_date.strftime("%Y-%m-%d")

        #inserting into latecoming table # 1 mark for inserting into Latecoming table correctly
        connection.execute("INSERT INTO Latecoming (StudentID, Latecoming_Date, Time_Arrived, Detention_Date, Detention_Completion_Status) VALUES (?,?,?,?,?)", (studentID, date_of_latecoming_str, time, detention_date_str, "False"))


    # Step 4 - make changes permanent
    connection.commit()

    # Step 5 - destroy database connection
    connection.close()
```
