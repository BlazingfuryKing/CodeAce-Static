```python
#Task 2.1
class Item:
    def __init__(self):
        self.__CatalogueNumber = -1 #INT
        self.__Brand = "" #STRING
        self.__Style = "" #STRING
        self.__Colour = ""#STRING
        self.__Rental = -1 #INT
        
    def getBrand(self): #return string
        return self.__Brand
    def setBrand(self,brand):
        self.__Brand = brand
    
    def getStyle(self):#return string
        return self.__Style
    def setStyle(self,style):
        self.__Style = style
    
    def getColour(self):#return string
        return self.__Colour
    def setColour(self,colour):
        self.__Colour = colour
    
    def getRental(self):#return integer
        return self.__Rental
    def setRental(self,rental):
        self.__Rental = rental
    
    def getCatalogueNumber(self):#return integer
        return self.__CatalogueNumber
    def setCatalogueNumber(self,catalogueNumber):
        self.__CatalogueNumber = catalogueNumber
        
        
class Apparel(Item):
    def __init__(self):
        super().__init__()
        self.__Size = -1 #INT
        self.__Fabrication = ""#STRING
        
    def getSize(self): #return integer
        return self.__Size
    def setSize(self,size):
        self.__Size = size
        
    def getFabrication(self):#return string
        return self.__Fabrication
    def setFabrication(self,fabrication):
        self.__Fabrication = fabrication
        
class Bag(Item):
    def __init__(self):
        super().__init__()
        self.__Length = -1 #INT
        self.__Breadth = -1 #INT
        self.__Leather = None #BOOLEAN
        
    def getLength(self): #return integer
        return self.__Length
    def setLength(self,length):
        self.__Length = length
    
    def getBreadth(self): #return integer
        return self.__Breadth
    def setBreadth(self,breadth):
        self.__Breadth = breadth
    
    def getLeather(self): #return boolean
        return self.__Leather
    def setLeather(self,leather):
        self.__Leather = leather
```

```python
#Task 2.2 
def Hashing(CatalogueNumber): #hashing algo
    return CatalogueNumber % 100

def Add_to_hashtable(ht,index,obj,email):
    while ht[index] != (): #while not tombstone value
        index = (index + 1) % 100
    
    ht[index] = (obj,email) #insert tuple
    return ht

hashtable = [() for i in range(100)] #hashtable of tuple
records_file = open("RECORDS.csv",'r')
records_file.readline() #header
#file processing
for line in records_file:
    
    CatalogueNumber,Type,Brand,Style,Colour,Rental,Size,Fabrication,Length,Breadth,Leather,Email = line.strip().split(",")
    
    CatalogueNumber = int(CatalogueNumber)
    Rental = int(Rental)
    
    #hash index
    index = Hashing(CatalogueNumber)
    
    #child class methods
    if Type == "Apparel":
        Size = int(Size)
        obj = Apparel()
        obj.setSize(Size)
        obj.setFabrication(Fabrication)
    
    else:
        Length = int(Length)
        Breadth = int(Breadth)
        Leather = Leather == "TRUE"
        obj = Bag()
        obj.setLength(Length)
        obj.setBreadth(Breadth)
        obj.setLeather(Leather)
    
    #parent class methods
    obj.setCatalogueNumber(CatalogueNumber)
    obj.setBrand(Brand)
    obj.setStyle(Style)
    obj.setColour(Colour)
    obj.setRental(Rental)
    
    #add to hashtable
    hashtable = Add_to_hashtable(hashtable,index,obj,Email)
    
records_file.close()
```

```python
#Task 2.3
def display(ht,start,end):
    print("{:10}{:20}{:20}".format("index","Catalogue Number","Email"))
    for index in range(start,end+1):
        if ht[index] == ():
            print("{:<10}{:20}{:20}".format(index,"",""))
        else:
            print("{:<10}{:<20}{:20}".format(index,ht[index][0].getCatalogueNumber(),ht[index][1]))
            
display(hashtable,0,9)
```

```python
#Task 2.4
def HashTableSearch(hashtable,CatalogueNumber):
    index = Hashing(CatalogueNumber)
    while hashtable[index] != (): #while not tomb stone
        
        if hashtable[index][0].getCatalogueNumber() == CatalogueNumber: 
            return hashtable[index][1]
        
        else: #continue traversing hashtable
            index = (index + 1) % 100
    
    #not found
    if hashtable[index] == ():
        return "Not Found"
```

```python
#Task 2.5
result1 = HashTableSearch(hashtable,1399)
result2 = HashTableSearch(hashtable,1220)
print(result1)
print(result2)
```
